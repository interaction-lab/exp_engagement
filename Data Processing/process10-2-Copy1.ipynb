```python
import pandas as pd
import numpy as np
from pandas import *
from scipy import stats
import os
import re
```

```python
path = '../../../../Google Drive File Stream/My Drive/USC Expeditions Year 5/Analysis/Help-Seeking/Data/p10_data2-1.csv'
data = pd.read_csv(path)
```

```python
# What are non open face/open pose columns? 
# Feature Dictionary Documentation here: https://docs.google.com/document/d/1RSygoLwsM1PKEIOoqDOaEo3lRBZBLZJuzbMSY5K_6zA/edit?usp=sharing

of_columns = []

for i in data.columns:
    # open face/open pose columns begin with of_ or op_
    if 'op' not in i and 'of' not in i:
        print(i)
    if 'of' in i:
        of_columns.append(i)
```

```python
# drop session 0

data = data[~(data['session_num'] == 0)]  
```

```python
answer = pd.DataFrame(index=[0,1], columns=data.columns)
```

```python

other_cols = ['games_total', 'games_session', 'mistakes_total', 'mistakes_session', 'mistakes_game', 'ts_robot_talked', 'ts_game_start', 'ts_attempt']

for c in data.columns:
    if 'op_' in c or 'of_' in c or c in other_cols:
        print(c)
        
        try: 
            min_val = data[c].min()
        except:
            data[c] = pd.to_numeric(data[c], errors='coerce')
            min_val = data[c].min()

        min_val = data[c].min()
        max_val = data[c].max()
        
        if pd.isnull(min_val):
            print('This column is null: ', c)
            answer[c] = 0.0
        else:
            answer.at[0, c] = min_val
            answer.at[1, c] = max_val
```

```python
path_data = '../../../../Google Drive File Stream/My Drive/USC Expeditions Year 5/Analysis/Help-Seeking/Data/p10_interm.csv'

answer.to_csv(path_data, index=False)
```
