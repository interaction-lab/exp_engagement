```python
import pandas as pd
import os
import regex as re
```

```python
participant = 18
data_path = '../../../../../Google Drive File Stream/My Drive/USC Expeditions Year 5/Analysis/Engagement/Data/ROS/ROS_Sessions/p'+str(participant) +'/Final'
out = '../../../../../Google Drive File Stream/My Drive/USC Expeditions Year 5/Analysis/Engagement/Data/ROS/p'+str(participant)+'_ros.csv'
```

```python
master = pd.DataFrame(columns=['participant', 'session_num', 'timestamp', 'ROBOT STATE', 'PARTICIPANT STATE', 'GAME STATE',
       'mistake_made', 'game_incorrect', 'game_correct', 'game_start',
       'ts_robot_talked', 'ts_game_start', 'ts_attempt', 'games_session',
       'mistakes_session', 'mistakes_game', 'activity', 'difficulty', 'skill',
       'aptitude', 'diff_1', 'diff_2', 'diff_3', 'diff_4', 'diff_5',
       'skill_EM', 'skill_NC', 'in_game'])
```

```python
master.head()
```

```python
for subdir, dirs, files in os.walk(data_path):
    for file in files:
        if 'ros' in file and 'csv' in file:
            path_to_file = data_path + '/' + file
            print(file)
            
            sess_regex1 = 's\d*_'
            sess_regex2 = 's\d*-\d*_'

            res1 = re.findall(sess_regex1, file)
            res2 = re.findall(sess_regex2, file)

            if (res1):
                session = res1[0]
                session = float(session[1:-1])
            else:
                session = res2[0]
                session = session[1:-1]
                session = session.replace('-', '.')
                session = float(session)
           
            process = pd.read_csv(path_to_file)
            process['participant'] = participant
            process['session_num'] = session
            
            master = pd.concat([master,process], sort=True)
```

```python
master = master.sort_values(by=['session_num', 'timestamp'])
```
