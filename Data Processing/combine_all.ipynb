```python
import pandas as pd
```

```python
p_path = 'p9'
ofp_df_path = '../../../../Google Drive File Stream/My Drive/USC Expeditions Year 5/Analysis/Engagement/Data/OF_P/'+ p_path + '_OF_P.csv'  
a_df_path = '../../../../Google Drive File Stream/My Drive/USC Expeditions Year 5/Analysis/Engagement/Data/Audio/'+ p_path + '_audio.csv'
e_df_path = '../../../../Google Drive File Stream/My Drive/USC Expeditions Year 5/Analysis/Engagement/Data/Annotations/'+ p_path +'_engagement.csv'
```

```python
ofp_df = pd.read_csv(ofp_df_path)
a_df = pd.read_csv(a_df_path)
e_df = pd.read_csv(e_df_path)
```

```python
# Only Keep Rows with Valid Participant, temporarily remove participant column
ofp_df = ofp_df.loc[ofp_df['participant'].notna()]
ofp_df = ofp_df.drop(columns=['participant'])
a_df = a_df.drop(columns=['participant'])
```

```python
# Only Keep Sessions with Annotations

keep =list(e_df['session_num'].unique())
a_df = a_df.loc[a_df['session_num'].isin(keep)]
ofp_df = ofp_df.loc[ofp_df['session_num'].isin(keep)]
```

```python
# Sort: session_num => timestamp
ofp_df = ofp_df.sort_values(by=['session_num', 'timestamp'])
a_df = a_df.sort_values(by=['session_num', 'timestamp'])
e_df = e_df.sort_values(by=['session_num', 'timestamp'])

ofp_df = ofp_df.reset_index(drop=True)
a_df = a_df.reset_index(drop=True)
e_df = e_df.reset_index(drop=True)
```

```python
e_df['timestamp'] = e_df['timestamp'].round(4)
ofp_df['timestamp'] = ofp_df['timestamp'].round(4)
a_df['timestamp'] = a_df['timestamp'].round(4)

e_df['session_num'] = e_df['session_num'].round(1)
ofp_df['session_num'] = ofp_df['session_num'].round(1)
a_df['session_num'] = a_df['session_num'].round(1)
```

```python
master = pd.merge(ofp_df, a_df, how='left', on=['session_num', 'timestamp'])
```

```python
master = pd.merge(master, e_df, how='left', on=['session_num', 'timestamp'])
```
