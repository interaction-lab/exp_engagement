```python
file = '../../../../Google Drive File Stream/My Drive/USC Expeditions Year 5/Analysis/Help-Seeking/Data/Affectiva Features/affectiva_webcam_2018-06-07-14-01-26.csv'
aff = pd.read_csv(file)
```

```python
aff.head()
```

```python
import decimal

def drange(x, y, jump):
  while x < y:
    yield float(x)
    x += decimal.Decimal(jump)
```

```python
percentile = list(drange(0, 100, '0.01'))
percentile.append(100)

percentile_track = []

for i in percentile:
    percentile_track.append(np.percentile(aff['Engagement'], i))

plt.plot(percentile, percentile_track)
plt.xlabel('Percentile')
plt.ylabel('Affectiva Engagement')
plt.title('Affectiva Engagement: Percentile Chart')
plt.show() 
```
