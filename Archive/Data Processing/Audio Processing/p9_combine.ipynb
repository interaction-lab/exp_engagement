```python
import os
import pandas as pd
import numpy as np
```

```python
directory = 'p9_audio'

df = pd.read_csv('p9_FS.csv')
```

```python
df['timestamp'] = df['timestamp'].round(6)
```

```python
df_audio = pd.DataFrame()
```

```python
for filename in os.listdir(directory):
    df_file = pd.read_csv(directory+'/'+filename)
    s1 = filename.split('_')[1]
    uval = np.float(s1.replace('s','').replace('-','.'))
    if(uval == 13.3):
        uval = np.float(13.0)
    df_file['timestamp'] = df_file.index * 0.03333333333333    
    df_file['timestamp'] = df_file['timestamp'].round(6)
    df_file['session_num'] = uval
    
    df_audio = pd.concat([df_audio,df_file],ignore_index=True)
```

```python
df_new = pd.merge(df, df_audio, how='left', on=['session_num', 'timestamp'])
```
