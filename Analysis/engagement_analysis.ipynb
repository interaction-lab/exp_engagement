```python
data = [data3, data5, data7, data8, data9, data10]
part = [3, 5, 7, 8, 9, 10]
```

```python
# Engagement By Participant 
height = []
height2 = []
for i in range(0, len(data)):
    counts = data[i]['engagement'].value_counts()
    #print(data[i]['engagement'].unique())
    #print(counts[1]+counts[0], len(data[i]))
    height.append((counts[1]/len(data[i]))*100)
    height2.append((counts[1]/len(data[i]))*100)

    #print(round((counts[1]/len(data[i]))*100, 3), end='%\n')
    #print(round(((counts[0]/len(data[i]))*100), 3), end='%\n')
    #print()

plt.rcParams["figure.figsize"] = [7,4]

fig, ax = plt.subplots()
index = np.arange(6)
bar_width = 0.6

err = [7.516, 15.215, 14.831, 23.719, 15.682, 23.521]
rects1 = ax.bar(index, height, bar_width, yerr=err)

ax.set_xlabel('Participant')
ax.set_ylabel('Engaged (%)')
ax.set_title('Engagement by Participant')

ax.set_xticks(index)
ax.set_xticklabels(('3', '5', '7', '8', '9', '10'))
fig.tight_layout()

plt.show()
```

```python
# Engagement By Session

engaged_map = {}
longest = 0
for i in range(0, len(data)-1):
    sessions = data[i]['session_num'].unique()
    sessions = sorted(sessions)
    engaged = []
    disengaged = []
    index = []
    for j,s in enumerate(sessions):
        index.append(j)
        curr = data[i].loc[data[i]['session_num'] == s]
        counts = curr['engagement'].value_counts()
        e = counts[1]/len(curr)
        d = counts[0]/len(curr)
        engaged.append(e*100)
        disengaged.append(d)
    engaged_map[i] = engaged
    longest = max(len(engaged), longest)
    # Variance of Engagement by Session 
    print(round((np.std(engaged))*100, 3), end='%\n')
    
    # Pearson Correlation Session, Engagement 
    #print(round(((sp.stats.pearsonr(index,engaged))[0]), 3))

    # Spearman Correlation Session, Engagement 
    #print(round(((sp.stats.spearmanr(index,engaged))[0]), 3))
    #print()

index = list(range(1, longest+1))

plt.rcParams["figure.figsize"] = [8,3]
for i,v in engaged_map.items():
    need = np.empty(longest-len(v))
    need[:] = np.nan
    new = np.hstack((np.array(v), need))
    #engaged_map[i] = new
    plt.plot(index, new, label=str(part[i]))

plt.legend(loc='best')
plt.xlabel('Session Index')
plt.ylabel('Engaged (%)')
plt.title('Engagement by Session')
```

```python
# Games Completed By Session

mistakes_map = {}
longest = 0
for i in range(0, len(data)-1):
    sessions = data[i]['session_num'].unique()
    sessions = sorted(sessions)
    mistakes = []
    index = []
    for j,s in enumerate(sessions):
        index.append(j)
        curr = data[i].loc[data[i]['session_num'] == s]
        mistakes.append(np.max(curr['games_session']))

    mistakes_map[i] = mistakes
    longest = max(len(mistakes), longest)
    # Variance of Engagement by Session 
    # print(round((np.std(engaged))*100, 3), end='%\n')
    
    # Pearson Correlation Session, Engagement 
    #print(round(((sp.stats.pearsonr(index,engaged))[0]), 3))

    # Spearman Correlation Session, Engagement 
    #print(round(((sp.stats.spearmanr(index,engaged))[0]), 3))
    #print()

index = list(range(1, longest+1))

plt.rcParams["figure.figsize"] = [8,3]
for i,v in mistakes_map.items():
    need = np.empty(longest-len(v))
    need[:] = np.nan
    new = np.hstack((np.array(v), need))
    #engaged_map[i] = new
    plt.plot(index, new, label=str(part[i]))

plt.legend(loc='best')
plt.xlabel('Session Index')
plt.ylabel('Games Played')
plt.title('Games Played by Session')
```

```python
# Continuous Engagement 

for i in range(0, len(data)):
    d = data[i]
    
    curr = 0
    last = d.loc[0, 'engagement']
    last_sess = d.loc[0, 'session_num']
    
    track_en = []
    track_dis = []
    
    for i,r in d.iterrows():
        if (last != r['engagement']) or (last_sess != r['session_num']):
            # record this 
            if (last == 0):
                track_dis.append(curr)
            else:
                track_en.append(curr)
            
            curr = 0
            last = r['engagement']
            last_sess = r['session_num']
        
        curr += 1
    
    # grab last one 
    if (last == 0):
        track_dis.append(curr)
    else:
        track_en.append(curr)
     
    # Median Continuous Engagement
    print(round((np.median(track_en)/30), 3))
    print(round((np.median(track_dis)/30), 3))
    
    # St Dev Continuous Engagement 
    print(round((np.std(track_en)/30), 3))
    print(round((np.std(track_dis)/30), 3))
    
    print()
```

```python
# Continuous Engagement Graph

plt.rcParams["figure.figsize"] = [7,4]

c_eng = [29.9, 24.767, 31.467, 20.833, 31.733, 7.3]

c_dis = [4.9, 6.267, 6.8, 11.117, 13.033, 5.767]

fig, ax = plt.subplots()

index = np.arange(6)
bar_width = 0.35

rects1 = ax.bar(index, c_eng, bar_width, label='engaged')
rects2 = ax.bar(index+bar_width, c_dis, bar_width, label='disengaged')

ax.set_xlabel('Participant')
ax.set_ylabel('Continuous Focus (Seconds)')
ax.set_title('Median Continuous Engagement and Disengagement')

ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(('3', '5', '7', '8', '9', '10'))
ax.legend()
fig.tight_layout()

plt.show()
```
