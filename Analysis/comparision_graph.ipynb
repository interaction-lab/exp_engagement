```python
import pandas as pd
import numpy as np
from scipy import stats 
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import regex as re
```

```python
# Modal Type 

gradient = ['#ffa600', '#ff764a', '#ef5675', '#bc5090', '#7a5195', '#374c80','#003f5c']
gradient.reverse()

gradient = ['#004c6d','#00678a','#0083a6','#00a1c1','#00c0d8','#00dfed','#00ffff']

models = ['Gradient Boosted Decision Trees (88%)', 'Random Decision Tree Forests (86%)', 'Logistic Regression (85%)', 'Neural Networks (82%)', 'Support Vector Machines (79%)', 'K-Nearest Neighbors (74%)', 'Naive Bayes (74%)']

num = 7
values = [88, 86, 85, 82, 82, 79, 74, 74]
startingRadius = 0.7 + (0.3* (num-1))

plt.rcParams["figure.figsize"] = [5.5, 4]

fig, ax = plt.subplots()

for index in range(num):
    percentage = values[index]
    remainingPie = 100-percentage

    donut_sizes = [remainingPie, percentage]

    #plt.text(0.01, startingRadius + 0.07, 's', horizontalalignment='center', verticalalignment='center')
    plt.pie(donut_sizes, radius=startingRadius, startangle=90, colors=['#ffffff', gradient[index]],
            wedgeprops={"edgecolor": "white"})

    startingRadius-=0.3

plt.title('Modal Type: AUROC (%)')
# equal ensures pie chart is drawn as a circle (equal aspect ratio)
plt.axis('equal')

# create circle and place onto pie chart
circle = plt.Circle(xy=(0, 0), radius=0.35, facecolor='white', edgecolor='white')
plt.gca().add_artist(circle)
plt.show()
fig.savefig('compare_models.eps', format='eps', dpi=1000)
```

```python
# Modal Type Legend

plt.rcParams["figure.figsize"] = [5.5, 4]

fig, ax = plt.subplots()

index = np.arange(1)
bar_width = 1/3


for i in range(num):
    rects1 = ax.bar(index, [0], bar_width, label=models[i], color=gradient[i])

ax.set_ylim(0, 125)



ax.legend(loc='center')
plt.show()
fig.savefig('modal_legend.eps', format='eps', dpi=1000)
```
